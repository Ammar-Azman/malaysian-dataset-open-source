```python
from glob import glob
import json

files = glob('sample.json_*.json')
files
```

```python
from tqdm import tqdm

data = []
for f in files:
    with open(f) as fopen:
        d = json.load(fopen)
        for d_ in tqdm(d):
            if d_ is None:
                continue
            
            success = False
            try:
                r = json.loads(d_[1])
                success = True
            except:
                pass
            
            if not success:
                try:
                    r = eval(d_[1])
                    success = True
                except:
                    pass
            
            if success:
                data.append((d_[0], r))
```

```python
import malaya

language_detection = malaya.language_detection.fasttext(model = 'mesolitica/fasttext-language-detection-bahasa-en')
```

```python
language_detection.predict(['ayam'])
```

```python
filtered = []
for d in data:
    try:
        ori = d[0]
        a = {}
        if isinstance(d[1], tuple):
            for d_ in d[1]:
                a = {**a, **d_}
        
        elif isinstance(d[1], set):
            for d_ in d[1]:
                l = language_detection.predict([d_])[0]
                if l == 'bahasa':
                    a['malay'] = d_
                elif l == 'english':
                    a['english'] = d_
        elif isinstance(d[1], str):
            l = language_detection.predict([d[1]])[0]
            if l == 'bahasa':
                a['malay'] = d[1]
            elif l == 'english':
                a['english'] = d[1]
        elif ori in d[1]:
            a = d[1][ori]
        else:
            a = d[1]
            
        for k in a:
            if isinstance(a[k], list):
                a[k] = '  '.join(a[k])
            
        if not isinstance(a.get('Malay', a.get('malay')), str) and not isinstance(a.get('English', a.get('english')), str):
            print(ori, 'aaaaa', a, '\n')
            continue

        filtered.append((ori, a))
    except Exception as e:
        # print(e, d, '\n')
        pass
```

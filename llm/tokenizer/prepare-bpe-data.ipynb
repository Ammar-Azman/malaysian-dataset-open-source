```python
files = [
    'c-cari-com-my.jsonl',
    'eprints.jsonl',
    'carigold.jsonl',
    'hardwarezone-sg.jsonl',
]
```

```python
from tqdm import tqdm
import json

with open('texts', 'w') as fopen_l:
    for f in files:
        count = 0
        with open(f) as fopen:
            for l in tqdm(fopen):
                data = json.loads(l)
                fopen_l.write(f'{data}\n')
                count += 1
                if count >= 1e6:
                    break
```

```python
from glob import glob

count = 0

with open('texts', 'a') as fopen_l:
    for f in glob('/home/husein/ssd3/google-translate-malay-news/*.requested'):
        
        if count >= 1e6:
            break
            
        with open(f) as fopen:
            for l in fopen:
                try:
                    data = json.loads(l)['r']['result']
                    fopen_l.write(f'{data}\n')
                    count += 1
                    
                except:
                    pass
                
                if count >= 1e6:
                    break
```

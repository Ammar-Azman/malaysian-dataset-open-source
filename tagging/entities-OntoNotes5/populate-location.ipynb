```python
pelancongan = [i.split(',')[0].strip() for i in pelancongan.split('\n') if len(i) > 0]
len(pelancongan)
```

```python
results = []
for d in pelancongan:
    d = d.strip()
    if len(d):
        entities = defaultdict(list)
        entities['tempat pelancongan'].append(d)
        entities['place'].append(d)
        results.append({'text': d, 'entities': entities})
```

```python
with open('LOC/gunung.txt') as fopen:
    data = fopen.read().split('\n')
    
gunung = []
for t in data:
    try:
        gunung.append(t.split('\t')[1])
    except:
        pass
len(gunung)
```

```python
for d in gunung:
    d = d.strip()
    if len(d):
        entities = defaultdict(list)
        entities['gunung'].append(d)
        entities['place'].append(d)
        results.append({'text': d, 'entities': entities})
```

```python
with open('LOC/planet.txt') as fopen:
    data = fopen.read().split('\n')
    
planet = []
for t in data:
    try:
        planet.append(t.split('(')[0].strip())
    except:
        pass
for d in planet:
    d = d.strip()
    if len(d):
        entities = defaultdict(list)
        entities['planet'].append(d)
        entities['place'].append(d)
        results.append({'text': d, 'entities': entities})
```

```python
with open('LOC/sungai.txt') as fopen:
    data = fopen.read().split('\n')
    
sungai = data[:]
sungai = list(filter(None, sungai))
for d in sungai:
    d = d.strip()
    if len(d):
        entities = defaultdict(list)
        entities['sungai'].append(d)
        entities['badan air'].append(d)
        entities['place'].append(d)
        results.append({'text': d, 'entities': entities})
```

```python
with open('LOC/tasik.txt') as fopen:
    data = fopen.read().split('\n')
    
tasik = data[:]
tasik = list(filter(None, tasik))
for d in tasik:
    d = d.strip()
    if len(d):
        entities = defaultdict(list)
        entities['tasik'].append(d)
        entities['badan air'].append(d)
        entities['place'].append(d)
        results.append({'text': d, 'entities': entities})
```

```python
import json

with open('augmentation-location.json', 'w') as fopen:
    json.dump(results, fopen)
```

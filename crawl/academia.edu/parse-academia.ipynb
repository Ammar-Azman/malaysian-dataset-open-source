```python
import dask.bag as db

def get_download(url):
    r = requests.get(url)
    bs = BeautifulSoup(r._content)
    try:
        return bs.find_all('a', {'class': 'js-swp-download-button'})[1].get('href')
    except Exception as e:
        print(url, e)
        return None

def get_nested(url):
    r = requests.get(url)
    bs = BeautifulSoup(r._content)
    suggested = bs.find_all('a', {'data-kale-collection': 'swp_suggested_papers'})
    bags = db.from_sequence([s.get('href') for s in suggested])
    mapped = bags.map(get_download)
    computed = mapped.compute(scheduler = 'threads', num_workers = 5)
    computed = list(filter(None, computed))
        
    return computed
```

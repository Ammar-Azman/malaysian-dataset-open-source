```python
from tqdm import tqdm

def generator():
    while True:
        yield
```

```python
result = []
i = 0
for _ in tqdm(generator()):
    params = {
        'no_definitions': 'true',
        'page': str(i),
    }

    response = requests.get('https://perak.org/latest.json', params=params, cookies=cookies, headers=headers)
    r = response.json()
    
    if not len(r['topic_list']['topics']):
        break
        
    result.append(r)
    
    i += 1
```

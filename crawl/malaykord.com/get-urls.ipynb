```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import os
import json
import numpy as np
import time
import random
import json
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
chars = 'qwertyuiopasdfghjklzxcvbnm'
all_a = []
for c in tqdm(chars):
    r = requests.get(f'https://www.malaykord.com/2014/03/list-chord-guitar-{c}.html')
    soup = BeautifulSoup(r.content, "lxml")
    a = soup.find_all('a', {'dir':"ltr"})
    a = [a_.get('href') for a_ in a]
    all_a.extend(a)
```

```python
all_a = sorted(list(set(all_a)))
len(all_a)
```

```python
all_a
```

```python
nested_a = []
for url in tqdm(all_a):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "lxml")
    for div in soup.find_all('div', {'class': 'post hentry uncustomized-post-template'}):
        for a in div.find_all('a'):
            href = a.get('href')
            if href:
                nested_a.append(href)
```

```python
all_a = []

for t in tqdm(topics):
    t = t.split('/')[-1]
    params = {
        'category': t,
        'filters[brand]': '',
        'filters[rating]': '',
        'filters[sortBy]': '',
        'page': '0',
        '_': '1689486712718',
    }

    response = requests.get(
        'https://www.tryandreview.com/my-en/result-search/category',
        params=params,
        cookies=cookies,
        headers=headers,
    )
    r = response.content
    soup = BeautifulSoup(r, "lxml")
    a = soup.find_all('a')
    a = [a_.get('href') for a_ in a]
    a = list(set(a))
    print(t, len(a))
    all_a.extend(a)
```

```python
all_a = list(set(all_a))
len(all_a)
```

```python
urls = [u.replace('#details', '').replace('#reviews', '') for u in all_a if '/my-en/brand/' not in u]
urls = sorted(list(set(urls)))
len(urls)
```

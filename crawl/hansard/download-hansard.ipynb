```python
import requests
import dateparser
from bs4 import BeautifulSoup
import os
```

```python
# download as HTML from https://www.parlimen.gov.my/hansard-dewan-rakyat.html?uweb=dr&arkib=yes

with open('Portal Rasmi Parlimen Malaysia - Penyata Rasmi.html') as fopen:
    data = fopen.read()
    
soup = BeautifulSoup(data, "lxml")
spans = soup.find_all("span", {"class": "standartTreeRow"})
dates = [s.text for s in spans]
len(spans)
```

```python
!mkdir pdf
```

```python
from tqdm import tqdm

filtered_date = []
for d in tqdm(dates):
    try:
        filtered_date.append(dateparser.parse(d).strftime('%d%m%Y'))
    except Exception as e:
        pass
```

```python
len(filtered_date)
```

```python
# https://www.parlimen.gov.my/files/hindex/pdf/DR-12102021.pdf
```

```python
def download(url, file_name):
    with open(file_name, "wb") as file:
        response = requests.get(url)
        file.write(response.content)
```

```python
for d in tqdm(filtered_date):
    filename = f'DR-{d}.pdf'
    output_filename = os.path.join('pdf', filename)
    if os.path.exists(output_filename):
        print(f'{output_filename} exists, skip.')
        continue
    url = f'https://www.parlimen.gov.my/files/hindex/pdf/{filename}'
    download(url, output_filename)
```

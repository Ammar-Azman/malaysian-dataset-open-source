```python
from tqdm import tqdm
import json

with open('news.json') as fopen:
    data = json.load(fopen)
```

```python
data[0]['topic_list']['topics'][0]['title']
```

```python
urls = []
for d in data:
    for t in d['topic_list']['topics']:
        title = t['title']
        id = t['id']
        url = f'https://perak.org/t/{title}/{id}'
        urls.append(url)
        
urls = sorted(list(set(urls)))
len(urls)
```

```python
urls
```

```python
def get_data(url):
    try:
        response = requests.get(url, cookies=cookies, headers=headers)
        return response.json()
    except Exception as e:
        print(e)
        return None
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
```

```python
# !rm -rf news
!mkdir news
```

```python
import os

max_worker = 3

for i in tqdm(range(0, len(urls), max_worker)):
    
    filename = os.path.join('news', f'{i}.json')
    if os.path.exists(filename):
        continue
    
    data = []    
    with ThreadPoolExecutor(max_workers=max_worker) as executor:
        futures = {executor.submit(get_data, url): url for url in urls[i: i + max_worker]}

        for future in as_completed(futures):
            r = future.result()
            if r:
                data.append(r)
            
    with open(filename, 'w') as fopen:
        json.dump(data, fopen)
```

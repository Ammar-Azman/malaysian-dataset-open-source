```python
years = []

for x in links:

    r = requests.get(x)
    soup = BeautifulSoup(r.content,'html.parser')

    container = soup.find('div', attrs = {'class':'js-tab-content-container'})

    year = [x.get('href') for x in container.find_all('a')]

    years.extend(year)
```

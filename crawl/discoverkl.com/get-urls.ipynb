```python
categories = [
    'audience/foodies',
    'audience/sports',
    'category/attractions',
    'category/events',
    'category/travel'
]
```

```python
import json
from tqdm import tqdm

all_a = []
for c in tqdm(categories):
    l, r = c.split('/')
    params = {
        'id': '',
        'post_id': '1779',
        'slug': 'sports',
        'canonical_url': 'https://discoverkl.com/audience/foodies/',
        'posts_per_page': '1000',
        'page': '0',
        'offset': '0',
        'post_type': 'post',
        'repeater': 'default',
        'seo_start_page': '1',
        'preloaded': 'false',
        'preloaded_amount': '100',
        'taxonomy': l,
        'taxonomy_terms': r,
        'taxonomy_operator': 'IN',
        'order': 'DESC',
        'orderby': 'date',
        'action': 'alm_get_posts',
        'query_type': 'standard',
    }

    response = requests.get('https://discoverkl.com/wp-admin/admin-ajax.php', params=params, cookies=cookies, headers=headers)
    r = response.json()
    soup = BeautifulSoup(r['html'], "lxml")
    a = soup.find_all('a', {'class':"post-list-title"})
    a = [a_.get('href') for a_ in a]
    all_a.extend(a)
```

```python
all_a = list(set(all_a))
```

```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
rs = []
for i in tqdm(range(1, 50, 1)):
    r = requests.get(f'https://www.ipendidikan.my/category/karangan/page/{i}')
    if r.status_code == 200:
        rs.append(r)
    else:
        break
```

```python
urls = []

for r_ in rs:
    soup = BeautifulSoup(r_.text, "lxml")
    for link in soup.find_all("a"):
        href = link.get("href")
        if 'semakan' in href or '.html' not in href or 'nota' in href or 'contoh' in href:
            continue
        urls.append(href)
        
urls = list(set(urls))
len(urls)
```

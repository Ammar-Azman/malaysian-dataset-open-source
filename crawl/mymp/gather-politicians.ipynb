```python
import json
import requests
from tqdm import tqdm
from bs4 import BeautifulSoup
```

```python
with open('politicians.json') as fopen:
    data = json.load(fopen)
```

```python
with open('MYMP.html') as fopen:
    html = fopen.read()
soup = BeautifulSoup(html, 'lxml')
```

```python
tags = soup.find_all('a')
tags = [t.attrs['href'].split('www.mymp.org.my/p/')[1] for t in tags if 'mymp.org.my/p/' in t.attrs['href']]
tags
```

```python
slugs = set(tags + [d['slug'] for d in data])
```

```python
datas = {}
for slug in tqdm(slugs):
    r = requests.get(f'https://mymp.org.my/p/{slug}')
    datas[slug] = r
```

```python
import pickle
```

```python
with open('mymp.pkl', 'wb') as fopen:
    pickle.dump(datas, fopen)
```

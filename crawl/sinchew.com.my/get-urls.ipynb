```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import os
import json
import numpy as np
import time
import random
import json
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
r = requests.get('https://www.sinchew.com.my/category/%e5%85%a8%e5%9b%bd/%e5%8d%b3%e6%97%b6%e5%9b%bd%e5%86%85/')
soup = BeautifulSoup(r.content, "lxml")
```

```python
a = soup.find_all('a')
a = [a_.get('href') for a_ in a if 'https://www.sinchew.com.my/category/' in a_.get('href')]
a = list(set(a))
a
```

```python
categories = []
for a_ in tqdm(a):
    r = requests.get(a_)
    categories.append(r.content.decode().split('data-catid="')[1].split('"')[0])
```

```python
categories = sorted(list(set(categories)))
```

```python
with open('categories.json', 'w') as fopen:
    json.dump(categories, fopen)
```

```python
def generator():
    while True:
        yield
    
for c in categories:
    index = 1
    for _ in tqdm(generator()):
        filename = os.path.join('url', f'{c}-{index}.json')
        if os.path.exists(filename):
            index += 1
            continue

        while True:
            try:
                r = requests.get(f'https://www.sinchew.com.my/ajx-api/category_posts/?cat={c}&page={index}&nooffset=true&editorialcat=0&posts_per_pages=10')
                break
            except:
                time.sleep(1.0)

        a = r.json()
        if not len(a):
            break

        with open(filename, 'w') as fopen:
            json.dump(a, fopen)

        index += 1
```

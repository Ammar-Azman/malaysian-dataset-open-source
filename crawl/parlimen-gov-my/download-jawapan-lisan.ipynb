```python
import requests
from bs4 import BeautifulSoup
import os
from tqdm import tqdm
from glob import glob
from datetime import datetime
import dateparser
from unidecode import unidecode
```

```python
# !pip3 install dateparser -U
```

```python
def download(url, file_name):
    with open(file_name, "wb") as file:
        response = requests.get(url)
        file.write(response.content)
```

```python
with open('Portal Rasmi Parlimen Malaysia - Jawapan Lisan.html') as fopen:
    data = fopen.read()
```

```python
soup = BeautifulSoup(data, "lxml")
```

```python
dateparser.parse(unidecode('05 April 2023'))
```

```python
date = []
for s in soup.find_all('span', {'class': 'standartTreeRow'}):
    try:
        date.append(dateparser.parse(s.text))
    except:
        pass
    
date = list(filter(None, date))
date = [d.strftime('%d%m%Y') for d in date]
date 
```

```python
!mkdir lisan
```

```python
for d in tqdm(date):
    url = f'http://parlimen.gov.my/files/jindex/pdf/JDR{d}.pdf'
    f = os.path.split(url)[1][-100:]
    f = os.path.join('lisan', f)
    if os.path.exists(f) and os.path.getsize(f) > 1000:
        continue
    download(url, f)
```

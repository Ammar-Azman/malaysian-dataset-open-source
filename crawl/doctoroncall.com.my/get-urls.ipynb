```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import os
import json
import numpy as np
import time
import random
import json
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
urls = [
    'https://www.doctoroncall.com.my/tanya/',
    'https://www.doctoroncall.com.my/tanya/en',
    'https://www.doctoroncall.com.my/tanya/zh',
]
```

```python
categories = []
for url in urls:
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "lxml")
    categories.extend(soup.find_all('a', {'class':"categories"}))
```

```python
categories = ['https://www.doctoroncall.com.my' + a_.get('href') for a_ in categories]
categories
```

```python
for c in categories:
    c_ = c.split('my/')[1].replace('/', '-').replace(' ', '-')
    r = requests.get(c)
    soup = BeautifulSoup(r.content, "lxml")
    nums = []
    for a in soup.find_all('a', {'class':"page-link"}):
        try:
            nums.append(int(a.text))
        except:
            pass

    if len(nums):
        page_num = max(nums)
    else:
        page_num = 1
        
    for index in tqdm(range(1, page_num + 1, 1)):
        filename = os.path.join('url', f'{c_}-{index}.json')
        if os.path.exists(filename):
            continue
            
        url = f'{c}/{index}'
        r = requests.get(url)
        soup = BeautifulSoup(r.content, "lxml")
        a = soup.find_all('a', {'class': 'topic-title topic_title_link'})
        a = [a_.get('href') for a_ in a]
        
        with open(filename, 'w') as fopen:
            json.dump(a, fopen)
```

```python
r = requests.get('https://www.e-mjm.org/past_issues.html', headers = headers, cookies = cookies)
```

```python
soup = BeautifulSoup(r.content)
a = soup.find_all('a')
pages = ['https://www.e-mjm.org/' + a_.get('href') for a_ in a if 'index.html' in a_.get('href')]
pages
```

```python
from tqdm import tqdm

urls = []
for p in tqdm(pages):
    r = requests.get(p, headers = headers, cookies = cookies)
    soup = BeautifulSoup(r.content)
    a = soup.find_all('a')
    a = ['https://www.e-mjm.org/' + a_.get('href') for a_ in a if a_.get('href', '').endswith('.pdf')]
    urls.extend(a)
    
urls
```

```python
urls_combined = sorted(list(set(urls)))
len(urls_combined)
```

```python
import os

pdfs = urls_combined
for i in tqdm(range(len(pdfs))):
    filename = os.path.join('e-mjm.org', f'{i}.pdf')
    if os.path.exists(filename):
        continue
        
    u = pdfs[i]
        
    r = requests.get(u, headers = headers, cookies = cookies)
    with open(filename, 'wb') as fopen:
        fopen.write(r.content)
```

```python
from glob import glob
from tika import parser
import json

TIKA_HOST = 'http://localhost:9998'
```

```python
files = glob('e-mjm.org/*.pdf')

with open('e-mjm.org.jsonl', 'w') as fopen:
    for f in tqdm(files):
        try:
            raw_xml = parser.from_file(f, TIKA_HOST, xmlContent=True)
            body = BeautifulSoup(raw_xml['content']).find('body')
            t = '\n'.join(body.findAll(text=True))
            fopen.write(f'{json.dumps(t)}\n')
        except:
            pass
```

```python
files = glob('topics/*.json')
```

```python
threads = []
for f in files:
    with open(f) as fopen:
        data = json.load(fopen)
    for d in data:
        splitted = d.split('-')
        splitted[-2] = '1'
        threads.append('-'.join(splitted))
        
threads = sorted(list(set(threads)))
len(threads)
```

```python
def get_url(url):
    while True:
        try:
            response = requests.get(url, cookies=cookies, headers=headers)
            soup = BeautifulSoup(response.content, "lxml")
            break
        except Exception as e:
            print(e)
            time.sleep(1.0)
    try:
        span = soup.find_all('span')
        max_page = [s.get('title') for s in span if '页' in s.get('title', '') and '共' in s.get('title', '')][0]
        max_page = int(max_page.split()[1])
    except:
        max_page = -1
    
    td = soup.find_all('td', {'class': 't_f'})
    td = [str(td_) for td_ in td]
    return url, td, max_page
```

```python
max_worker = 30
for i in tqdm(range(0, len(threads), max_worker)):
    filename = os.path.join('threads', f'{i}.json')
    if os.path.exists(filename):
        continue
    
    gc.collect()
    
    urls = [f'https://www.jbtalks.cc/{t}' for t in threads[i: i + max_worker]]
    results = []
    with ThreadPoolExecutor(max_workers=max_worker) as executor:
        futures = {executor.submit(get_url, url): url for url in urls}

        for future in as_completed(futures):
            results.append(future.result())
    
    with open(filename, 'w') as fopen:
        json.dump(results, fopen)
        
    try:
        del results
        del futures
    except:
        pass
```

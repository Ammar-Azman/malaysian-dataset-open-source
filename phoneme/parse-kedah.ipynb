```python
df = pd.read_csv('kedah.csv')
df
```

```python
df.iloc[0,1]
```

```python
import malaya
```

```python
columns = [1, 4, 5]
```

```python
results = {}
for c in columns:
    for i in range(len(df)):
        try:
            stripped = df.iloc[i, c].strip()
            if stripped[0] == '[' and stripped[-1] == ']':
                before = df.iloc[i, c - 1].strip()
                splitted = malaya.text.function.split_into_sentences(before, minimum_length = 4)
                if len(splitted) > 1:
                    stripped_splitted = stripped.split('] [')[0]
                    before = splitted[0]
                    stripped = stripped_splitted + ']'
                results[before] = stripped[1:-1]
        except Exception as e:
            print(e)
```

```python
results
```

```python
results['Hang tok sah dok ajak diala, saheh dia tak mau pi'] = \
'hAN to/ sAh do/ A.dZA/ di.jA.lA sA.hEh di.jA tA/ mAw pi'
```

```python
results['Mak buboh adan depan pintu tak bagi adek keluaq'] = \
'mA/ bu.boh A.dAn d«.pAn pin.tu tA/ bA.gi A.de/ k«.lu.wAâÖ'
```

```python
del results['Hang tok sah dok ajak diala, saheh dia tak mau pi pasai hati dia ahia.']
del results['Mak buboh adan depan pintu tak bagi adek keluaq rumah.']
```

```python
with open('kedah-phoneme.json', 'w') as fopen:
    json.dump(results, fopen)
```

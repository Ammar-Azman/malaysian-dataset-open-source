```python
import json
from tqdm import tqdm
```

```python
bahasa = []

with open('raw-wiktextract-data.json',  encoding="utf-8") as f:
    for line in tqdm(f):
        data = json.loads(line)
        lang = data.get('lang', '').lower()
        if 'malay' in lang or 'indonesia' in lang:
            bahasa.append(data)
```

```python
len(bahasa)
```

```python
selected_bahasa = [l for l in bahasa if l['lang'] not in ['Malayalam']]
len(selected_bahasa)
```

```python
[l for l in selected_bahasa if 'ayam' in l['word']]
```

```python
with open('wiktionary-bahasa.json', 'w') as fopen:
    json.dump(selected_bahasa, fopen)
```

```python
import redis
import json
from tqdm import tqdm
```

```python
db = redis.Redis(host='localhost', port=6379, db=10)
```

```python
keys = sorted(db.keys())
len(keys)
```

```python
keys[:1000]
```

```python
json.loads(db['aaau'])
```

```python
filtered = {}
for k in tqdm(keys):
    r = json.loads(db[k])
    if len(r['definisi']):
        filtered[k.decode()] = r
```

```python
len(filtered)
```

```python
filtered['terserlah']
```

```python
with open('dbp-v2.json', 'w') as fopen:
    json.dump(filtered, fopen)
```

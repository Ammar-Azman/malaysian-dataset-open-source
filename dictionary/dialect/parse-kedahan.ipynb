```python
import re
from unidecode import unidecode

def cleaning(string):
    string = unidecode(string).replace('\n', ' ').replace('\t', ' ')
    string = re.sub(r'[ ]+', ' ', string).strip()
    return string.lower()
```

```python
kedah = {}

for l in x.split('\n'):
    if not len(l):
        continue
    splitted = l.split('–')
    if not len(splitted) == 2:
        continue
    l = cleaning(splitted[0])
    r = cleaning(splitted[1])
    kedah[l] = [r]
```

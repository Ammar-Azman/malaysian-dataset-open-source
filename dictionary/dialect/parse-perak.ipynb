```python
import re
from unidecode import unidecode

def cleaning(string):
    string = unidecode(string).replace('\n', ' ').replace('\t', ' ')
    string = re.sub(r'[ ]+', ' ', string).strip()
    return string.lower()
```

```python
vocab = {}
for r in x.split('\n'):
    splitted = r.split(' = ')
    if not len(splitted) == 2:
        continue
    l = cleaning(splitted[0].replace('- ', ''))
    r = cleaning(splitted[1])
    vocab[l] = [r]
```

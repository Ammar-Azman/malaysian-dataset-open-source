```python
import json
from glob import glob
```

```python
files = glob('data/*.json')
len(files)
```

```python
replace = {
    'CIPTA JADUAL': 'CREATE TABLE',
}
```

```python
splitted = [
    'Penyelesaian:',
    'JSON struktur',
    'Untuk mengembalikan',
    'ke dalam Bahasa Melayu',
    'JSON structure',
    'Solusi:'
]
```

```python
mapping = {}
for f in files:
    with open(f) as fopen:
        try:
            d = json.load(fopen)
            try:
                malay = json.loads(d[1])['malay']
                
            except Exception as e:
                malay = d[1]
                for s in splitted:
                    if s in d[1]:
                        malay = malay.split(s)[1].strip()
                    
                if '{"malay' in malay:
                    malay = malay.replace('{"malay', '')
                    
                    
                if malay.endswith('"}'):
                    malay = malay[:-2]
                if malay.endswith('"\n}'):
                    malay = malay[:-3]
                    
                if '"malay"' in malay:
                    continue
                    
                if 'butuh' in malay:
                    continue

            for k, v in replace.items():
                malay = malay.replace(k, v)
                
            if (len(malay) / len(d[0])) < 0.7:
                continue
                
            mapping[d[0]] = malay
            
        except Exception as e:
            pass
            print(e)
```

```python
with open('combine.translated.jsonl', 'w') as fopen_l:
    with open('combine.jsonl') as fopen:
        for l in fopen:
            d = json.loads(l)
            d['instruction_ms'] = mapping.get(d['instruction'])
            d['output_ms'] = mapping.get(d['output'])
            fopen_l.write(f'{json.dumps(d)}\n')
```
